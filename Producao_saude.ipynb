```python
import pandas as pd
import matplotlib.pyplot as plt

dados = pd.read_csv('/content/A152918189_28_143_208.csv', encoding='ISO-8859-1', skiprows=3, sep=';', skipfooter=12, thousands='.', decimal=',')

# define as dimensões da figura
plt.figure(figsize=(20, 6))

# gera o gráfico em barras
plt.bar(dados["Unidade da Federação"], dados["2008/Ago"], width=2)

# rotaciona as legendas do eixo X em 45 graus
plt.xticks(rotation=45)

# adiciona o título do gráfico
plt.title('Título do gráfico')

# adiciona as legendas
plt.legend(loc='upper left', fontsize=10)

# exibe o gráfico
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

dados = pd.read_csv('/content/A152918189_28_143_208.csv', encoding='ISO-8859-1', skiprows=3, sep=';', skipfooter=12, thousands='.', decimal=',')

# define as dimensões da figura
plt.figure(figsize=(20, 6))

# gera o gráfico em barras
plt.barh(dados["Unidade da Federação"], dados["2008/Ago"])


# rotaciona as legendas do eixo X em 45 graus
plt.xticks(rotation=45)

# adiciona o título do gráfico
plt.title('Valor por unidade da federação')

# adiciona as legendas
plt.legend(loc='upper left', fontsize=10)

# exibe o gráfico
plt.show()
```
